## HOURLY

### FIND ADRESESS

```python
adresses = driver.find_elements(By.CSS_SELECTOR, "span[class='chakra-text FacilitySummary-title css-18lhgdc']")
for adress in adresses:
    print(adress.text)
```

```python
adresses_text=[]
for i in adresses:
     adresses_text.append(i.text)
```

```python
print(adresses_text)
```

```python
import pandas as pd
```

```python
adresses_df = pd.DataFrame(adresses_text, columns=['adresses'])
```

```python
print(adresses_df)
```

```python
adresses_df2 = pd.DataFrame(adresses_df.adresses.str.split('-',1).tolist(),
                                 columns = ['adresses','names'])
```

### FIND STARS

```python
stars = driver.find_elements(By.CSS_SELECTOR, "div[class='StarRatingWithCategory'] span")
for star in stars:
    print(star.text)
```

```python
stars_text=[]
for i in stars:
     stars_text.append(i.text)
```

```python
print(stars_text)
```

```python
stars_df = pd.DataFrame(stars_text, columns=['stars'])
```

### FIND PRICES

```python
prices = driver.find_elements(By.CSS_SELECTOR, "div[class='price']")
for price in prices:
    print(price.text)
```

```python
prices_text=[]
for i in prices:
     prices_text.append(i.text)
```

```python
prices_df = pd.DataFrame(prices_text, columns=['prices'])
```

```python
print(prices_df)
```

### FIND ALL

```python
data_spots = driver.find_elements(By.CLASS_NAME, "div[class='SpotList-spots']")
for data_spot in data_spots:
    print(data_spot.text)
```

```python
data_spots = driver.find_elements(By.XPATH, "//*[@id='root']/div/div[2]/div/div[2]/div[1]/div[3]/div")
for data_spot in data_spots:
    print(data_spot.text)
```

### FIND RESULT NAMES

```python
element = driver.find_element(By.XPATH, "//*[@id='root']/div/div[2]/div/div[2]/div[1]/div[3]/div")
children = element.find_elements(By.XPATH, "./child::*")
for child in children:
    idVal = child.get_attribute('class')
    print(idVal)
```

### FIND RESULT IDS

```python
mylist = []
element_ = driver.find_element(By.XPATH, "//*[@id='root']/div/div[2]/div/div[2]/div[1]/div[3]/div")
children_ = element_.find_elements(By.XPATH, "./child::*")
for child_ in children_:
    idVal_ = child_.get_attribute('data-spot')
    mylist.append(str(idVal_))
print(mylist)
```

```python
spot_df = pd.DataFrame(mylist, columns=['spot_id'])
```

```python
print(spot_df)
```

### TAKE ALL AS A DATA FRAME

```python
result_df = pd.concat([stars_df, prices_df, spot_df, adresses_df2], axis=1, join='outer')
```

```python
print(result_df)
```

```python
result_df.to_csv('hourly_spots2.csv')
```
